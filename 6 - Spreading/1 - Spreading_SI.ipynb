# ---- SI: Susceptible-Infected 

```python
output_notebook()
```

```python
beta = 0.001
percentage_infected = 0.05
```

```python
def start_SI(G, beta, percentage_infected, it_brunch):
    print(f"Usati beta = {beta}, percentage_infected = {percentage_infected}, it_brunch = {it_brunch}")
    config = mc.Configuration()
    config.add_model_parameter('beta', beta)
    config.add_model_parameter("percentage_infected", percentage_infected)
    model_G = ep.SIModel(G)
    model_G.set_initial_status(config)
    iterations_G = model_G.iteration_bunch(it_brunch, node_status=True)
    trends_G = model_G.build_trends(iterations_G)
    
    vm = MultiPlot()
    
    viz = DiffusionTrend(model_G, trends_G)
    p = viz.plot()

    vm.add_plot(p)
    
    viz2 = DiffusionPrevalence(model_G, trends_G)
    p2 = viz2.plot()

    
    vm.add_plot(p2)
    m = vm.plot()
    show(m)
```

```python
start_SI(G, beta, percentage_infected, 400)
```

```python
start_SI(BA, beta, percentage_infected, 400)
```

```python
start_SI(ER, beta, percentage_infected, 400)
```

```python
config = mc.Configuration()
config.add_model_parameter('beta', beta)
config.add_model_parameter("percentage_infected", percentage_infected)
model_G = ep.SIModel(ER)
model_G.set_initial_status(config)
iterations_G = model_G.iteration_bunch(400, node_status=True)
```

```python
for i in iterations_G:
    print(i["iteration"])
    print(i["node_count"])
        
```
