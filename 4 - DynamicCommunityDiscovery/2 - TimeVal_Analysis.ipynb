```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
path_tags= "../DataSet StackOverflow/tags_with_time.csv"
```

```python
path_couple= "../DataSet StackOverflow/tags_edges_with_time.csv"
```

```python
path_question = "../DataSet StackOverflow/question_with_time.csv"
```

```python
df_tags = pd.read_csv(path_tags)
```

```python
df_couple = pd.read_csv(path_couple)
```

```python
df_question = pd.read_csv(path_question)
```

```python
df_question["Id"] = [1 for i in range(len(df_question))]
```

```python

plt.plot([i for i in range(1, 13)], list(df_question.groupby(['Month']).sum()["Id"]) , color='blue', marker='o')
plt.title('Questions per Month', fontsize=14)
plt.xlabel('Month', fontsize=14)
plt.xticks(np.arange(13),["", "January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"], rotation=50)
plt.ylabel('Number of questions', fontsize=14)
plt.grid(True)
plt.show()
```
